```python
#imports
    
import matplotlib.pyplot as plt
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import random
import datetime as dt
import matplotlib.pyplot as plt
import datetime
from subprocess import check_output
```

```python
import os #change to data
os.chdir('../stockData')
#read in datafiles
filenames = [x for x in os.listdir('../stockData') if x.endswith('.txt') and os.path.getsize(x) > 0]

#random samples to check
#filenames = random.sample(filenames,10)
print(filenames)
```

```python
### we need to get datetimes for the input vectors. We are training three neural networks for this.
```

```python
data = []
for filename in filenames:
    df = pd.read_csv(filename, sep=',')
    #average price
    df['Price'] = (df['High'] + df['Low'] + df['Open'] + df['Close'])/4
    # do more feature engineering here:
    columns2Drop = ['High', 'Low', 'Open', 'Close', 'OpenInt']
    df = df.drop(labels=columns2Drop, axis=1)
   
    #create "movement feature"
    df['Volume*Price'] = df['Volume'] * df['Price'] 
    # create label
    
    label, _, _ = filename.split('.')
    df['Label'] = label
    df['Date'] = pd.to_datetime(df['Date'])

    data.append(df)
df = pd.concat(data)
print df.head()

df.to_csv('toGoogle.csv')    
```

```python
def createVector(startday,label,numdays):
    #takes in datetime object startday, which is the day to start by. 
    #takes in label so we know what stock to look for
    #takes in numdays to find out how many days we want to go back    
    inputVector=[]
    location = df.iloc[df["Date"].isoformat() == startday && df['Label']==label]
    timedelt = datetime.timedelta(days=numdays)
    startday
    inputVector.append(df.loc[df["Date"].isoformat() == startday] : timedelt, ["Volume", "Price","Volume*Price"]
    print(inputVector)  
        
    return inputVector
```

```python
def  getDayLists(ticker,day=7):#day better be a int
    # get start days
        backToPast = 344 #days since last recorded stock 
        day = day+backToPast
        startWeekDelta = datetime.timedelta(days=day, weeks=1)
        startWeekDay = (day- startWeekDelta).isoformat()
        startMonthDelta = datetime.timedelta(days=day, weeks=4)
        startMonthDay = (day - startMonthDelta).isoformat()
        startSixMonthDelta = datetime.timedelta(days=day, weeks=26)
        startSixMonth = (day - startSixMonthDelta).isoformat()
        #build vectors
        
        weekToAvg=createVector(startWeekDay,ticker,7)
        monthToAvg= createVector(startMonthDay,ticker,30)
        sixToAvg = createVector(startSixMonth, ticker, 182)
        
        return [weekToAvg, monthToAvg, sixToAvg]
    
    #we find the row number of the start day, then increment the index by the length of time(week, month, 6month) to get the rows
    
    
```
